### 2. Plotar os dados do primeiro sensor

```python
import numpy as np
import matplotlib.pyplot as plt

d1 = np.load('./Dados_Avaliação_P1_2024_1/Dados_1.npy')
tempo = np.arange(50000)/10000

plt.plot(tempo,d1[0:50000])
plt.xlabel('Segundos')
plt.ylabel('Medições obtidas')
plt.title('200 medições do SENSOR 1')
```

### 3. Plotagem dos dados de todos os SENSORES

```python
d2 = np.load('./Dados_Avaliação_P1_2024_1/Dados_2.npy')
d3 = np.load('./Dados_Avaliação_P1_2024_1/Dados_3.npy')
d4 = np.load('./Dados_Avaliação_P1_2024_1/Dados_4.npy')
d5 = np.load('./Dados_Avaliação_P1_2024_1/Dados_5.npy')

figure, axis = plt.subplots(3, 2) 
axis[0, 0].plot(tempo, d1[0:50000,0]); axis[0, 0].set_title("Sensor 1")
axis[0, 1].plot(tempo, d2[0:50000,0],color='r'); axis[0, 1].set_title("Sensor 2")
axis[1, 0].plot(tempo, d3[0:50000,0],color='g'); axis[1, 0].set_title("Sensor 3")
axis[1, 1].plot(tempo, d4[0:50000,0],color='m'); axis[1, 1].set_title("Sensor 4")
axis[2, 0].plot(tempo, d5[0:50000,0],color='k'); axis[2, 0].set_title("Sensor 5")
```

### 4. Diagrama de caixa da leitura dos sensores

```python
import matplotlib.pyplot as plt
import numpy as np

all_data = [d1[0:50000,0], d2[0:50000,0], d3[0:50000,0]]
labels = ['Sensor 1', 'Sensor 2', 'Sensor 3']

# dá para fazer uma estatística
print('Média (Sensor 1):', d1[0:50000,0].mean())
print('Desvio padrão (Sensor 1):', d1[0:50000,0].std())

# rectangular box plot
fig, ax = plt.subplots()
ax.boxplot(all_data, notch=True, vert=True, patch_artist=True, labels=labels)
plt.show()
```

### Conclusões: Temos 5 segundos de gravação de 5 sensores que possuem 200 dados de entrada. Em geral, os motores são monitorados em relação a tensão, corrente e temperatura. Os sensores 1, 2 e 3 possuem leituras simètricas. O sensor 4 parece ser de temperatura, e o sensor 5 possui leitura assimêtrica.
